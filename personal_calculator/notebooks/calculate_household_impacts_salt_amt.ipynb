```python
# Define the states and income levels we want to test
states = ["NY", "CA", "PA", "NJ"]
income_levels = [100000, 250000, 500000, 1000000]

# Create a list to store all results
all_results = []

# Loop through each state and income combination
for state in states:
    for income in income_levels:
        # Create the situation dictionary with this state and income
        situation = {
            "people": {
                "you": {
                    "age": {2026: 40},
                    "deductible_mortgage_interest": {2026: 15000},
                    "employment_income": {2026: income},
                    "charitable_cash_donations": {2026: 10000},
                },
                "your partner": {"age": {2026: 40}},
            },
            "families": {
                "your family": {
                    "members": [
                        "you",
                        "your partner",
                    ]
                }
            },
            "tax_units": {
                "your tax unit": {
                    "members": [
                        "you",
                        "your partner",
                    ]
                }
            },
            "spm_units": {
                "your household": {
                    "members": [
                        "you",
                        "your partner",
                    ]
                }
            },
            "households": {
                "your household": {
                    "members": [
                        "you",
                        "your partner",
                    ],
                    "state_name": {2026: state},
                }
            },
            # Keep the original axes configuration for property taxes
            "axes": [[{"name": "real_estate_taxes", "count": 500, "min": 0, "max": 50000}]],
        }
        
        # Run the simulations
        current_law_simulation = Simulation(situation=situation)
        current_policy_simulation = Simulation(situation=situation, reform=reform)
        
        # Calculate the metrics
        results = {}
        for simulation, name in [(current_law_simulation, "current_law"), 
                                (current_policy_simulation, "current_policy")]:
            results[name] = {
                "regular_tax_before_credits": simulation.calculate("regular_tax_before_credits", 2026),
                "tentative_minimum_tax": simulation.calculate("amt_base_tax", 2026),
                "salt_deduction": simulation.calculate("salt_deduction", 2026),
                "federal_tax": simulation.calculate("income_tax", 2026),
            }
        
        # Create a DataFrame for this state and income combination
        df = pd.DataFrame({
            "state": state,
            "income": income,
            "real_estate_taxes": np.arange(0, 50000, 100),  # Property taxes from $0 to $50k in $100 increments
            "current_law_regular_tax": results["current_law"]["regular_tax_before_credits"],
            "current_law_amt": results["current_law"]["tentative_minimum_tax"],
            "current_law_salt_deduction": results["current_law"]["salt_deduction"],
            "current_law_federal_tax": results["current_law"]["federal_tax"],
            "current_policy_regular_tax": results["current_policy"]["regular_tax_before_credits"],
            "current_policy_amt": results["current_policy"]["tentative_minimum_tax"],
            "current_policy_salt_deduction": results["current_policy"]["salt_deduction"],
            "current_policy_federal_tax": results["current_policy"]["federal_tax"],
        })
        
        # Add to our list of results
        all_results.append(df)

# Combine all results into a single DataFrame
combined_results = pd.concat(all_results, ignore_index=True)

# Save to CSV
combined_results.to_csv("../data/all_state_income_tax_calculations_2026.csv", index=False)

# Display the first few rows
combined_results.head()
```
