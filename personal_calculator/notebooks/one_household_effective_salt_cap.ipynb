```python
def create_situation_with_two_axes(income_value, has_spouse=True):
    """
    Create situation dictionary with two axes (reported_salt and fixed employment_income)
    
    Parameters:
    income_value: The fixed income value to use
    has_spouse: Boolean indicating whether to include a spouse in the household
    """
    if has_spouse:
        # Create a situation with a spouse
        situation = {
            "people": {
                "you": {"age": {"2026": 40},
                        "employment_income": {"2026": income_value}},
                "spouse": {"age": {"2026": 40}}
            },
            "families": {"your family": {"members": ["you", "spouse"]}},
            "marital_units": {"your marital unit": {"members": ["you", "spouse"]}},
            "tax_units": {
                "your tax unit": {
                    "members": ["you", "spouse"],
                    "state_and_local_sales_or_income_tax": {"2026": 0}
                }
            },
            "spm_units": {"your household": {"members": ["you", "spouse"]}},
            "households": {
                "your household": {
                    "members": ["you", "spouse"],
                    "state_code": {"2026": "CA"}
                }
            },
            "axes": [
                [
                    {
                        "name": "reported_salt",
                        "count": 700,
                        "min": -50000,
                        "max": 250000,
                        "period": 2026,
                    }
                ],
            ],
        }
    else:
        # Create a situation without a spouse (single filer)
        situation = {
            "people": {
                "you": {"age": {"2026": 40},
                        "employment_income": {"2026": income_value}}
            },
            "families": {"your family": {"members": ["you"]}},
            "marital_units": {"your marital unit": {"members": ["you"]}},
            "tax_units": {
                "your tax unit": {
                    "members": ["you"],
                    "state_and_local_sales_or_income_tax": {"2026": 0}
                }
            },
            "spm_units": {"your household": {"members": ["you"]}},
            "households": {
                "your household": {
                    "members": ["you"],
                    "state_code": {"2026": "CA"}
                }
            },
            "axes": [
                [
                    {
                        "name": "reported_salt",
                        "count": 700,
                        "min": -50000,
                        "max": 250000,
                        "period": 2026,
                    }
                ],
            ],
        }
    
    return situation
```

```python
def calculate_effective_salt_cap_for_income(income_value, has_spouse=True):
    """
    Calculate effective SALT cap for a specific income level
    
    Parameters:
    income_value: The fixed income value to use
    has_spouse: Boolean indicating whether to include a spouse in the household
    """
    # Create situation with the given income value and spouse setting
    situation = create_situation_with_two_axes(income_value, has_spouse)
    
    # Create simulation 
    simulation = Simulation(situation=situation, reform=tcja_reform)

    # Calculate values along the axes
    reported_salt = simulation.calculate("reported_salt", map_to="household", period=2026)
    income_tax = simulation.calculate("income_tax", map_to="household", period=2026)
    salt_deduction = simulation.calculate("salt_deduction", map_to="household", period=2026)

    # Create DataFrame with data
    df = pd.DataFrame({
        "employment_income": income_value,  # Use the fixed income value 
        "reported_salt": reported_salt,
        "income_tax": income_tax,
        "salt_deduction": salt_deduction,
    })
    
    return df
```

```python
def calculate_effective_salt_cap_over_earnings(has_spouse=True):
    """
    Calculate effective SALT cap for varying property tax and all income levels
    
    Parameters:
    has_spouse: Boolean indicating whether to include a spouse in the household
    """
    # Create an empty list to store DataFrames for each income level
    all_dfs = []
    
    # Loop through each income level
    for income in INCOMES:
        # Calculate for this specific income level with specified spouse setting
        df = calculate_effective_salt_cap_for_income(income, has_spouse)
        # Append to our list
        all_dfs.append(df)
    
    # Combine all DataFrames
    combined_df = pd.concat(all_dfs, ignore_index=True)
    
    return combined_df
```

```python
def calculate_marginal_rate(group):
    """Calculate the marginal property tax rate for a group"""
    # Check if group needs sorting
    if not group["reported_salt"].is_monotonic_increasing:
        group = group.sort_values("reported_salt")

    # Calculate differences in income tax and property tax
    group["income_tax_diff"] = group["income_tax"].diff()
    group["reported_salt_diff"] = group["reported_salt"].diff()

    # Calculate the marginal rate (change in income tax / change in property tax)
    group["marginal_property_tax_rate"] = -group["income_tax_diff"] / group["reported_salt_diff"]

    # Handle edge cases
    group["marginal_property_tax_rate"] = group["marginal_property_tax_rate"].replace(
        [np.inf, -np.inf, np.nan], 0
    )
    
    return group
```

```python
def process_effective_cap_data(df):
    """Process the data by calculating marginal rates for each income level"""
    # Sort the dataframe
    sorted_df = df.sort_values(by=["employment_income", "reported_salt"])
    
    # Calculate the marginal property tax rate for each income level
    processed_df = (
        sorted_df.groupby(["employment_income"])
        .apply(calculate_marginal_rate)
        .reset_index(drop=True)
    )
    
    return processed_df
```

```python
def get_effective_salt_caps(has_spouse=True, threshold=0.1):
    """
    Calculate and return the effective SALT cap by income level without plotting
    
    Parameters:
    has_spouse: Boolean indicating whether to include a spouse in the household
    threshold: The marginal property tax rate threshold to use for determining the effective cap
    """
    # Calculate data for all income levels with specified spouse setting
    result_df = calculate_effective_salt_cap_over_earnings(has_spouse)
    
    # Process the data to calculate marginal rates
    processed_df = process_effective_cap_data(result_df)
    
    # Filter data where marginal_property_tax_rate > threshold
    filtered_df = processed_df[processed_df["marginal_property_tax_rate"] > threshold]
    
    # For each unique employment income, find the maximum SALT value
    max_salt_by_income = (
        filtered_df.groupby("employment_income")["reported_salt"].max().reset_index()
    )
    
    # Sort by income for proper display
    max_salt_by_income = max_salt_by_income.sort_values("employment_income")
    
    # Format as a dictionary for easy access
    result = {}
    for _, row in max_salt_by_income.iterrows():
        income = row["employment_income"]
        salt_cap = row["reported_salt"]
        result[income] = salt_cap
    
    return result
```
